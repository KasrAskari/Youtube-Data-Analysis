# YouTube Data Analysis

### Import Libraries

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

### Reading Data

```python
df = pd.read_csv("youtube.csv", encoding="latin1")
df
```

### Grouping

```python
grouped_by_country = df.groupby("Country")["subscribers"].sum().sort_values(ascending=False)
grouped_by_country
```

### Charting

```python
plt.figure(figsize=(15,6))
grouped_by_country.plot(kind="bar")

plt.title("Total Subscribers Based on Countries")
plt.ylabel("subscribers")
plt.xlabel("country")

plt.show()
```

```python
top_10_categories = df["category"].value_counts().head(10)

plt.figure(figsize=(10,10))

plt.pie(top_10_categories, labels=top_10_categories.index, autopct="%1.1f%%", startangle=90)

plt.title("Top 10 categories")
plt.show()
```
